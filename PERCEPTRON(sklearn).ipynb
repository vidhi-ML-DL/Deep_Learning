```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('studentplacement_dataset(perceptron).csv')
```

```python
sns.scatterplot(data=df,x='CGPA',y='Resume_Score',hue = 'Placed')
```

```python
X = df.iloc[:,0:2]
Y = df.iloc[:,-1]
```

```python
p = Perceptron()
p.fit(X,Y)
coeff = p.coef_
intercept = p.intercept_

slope = -(coeff[0][0]/coeff[0][1])
b = -(intercept/coeff[0][1])

print(coeff,intercept,slope,b)
```

```python
predictions = p.predict(X)
accuracy = accuracy_score(Y,predictions)
print(accuracy)
```

```python
plt.figure(figsize=(10,6))
x_lim = np.linspace(6,10,100)
y_line = slope* x_lim + b
plt.plot(x_lim,y_line,color = 'Green',linewidth = 2)
plt.scatter(df['CGPA'], df['Resume_Score'], c=Y, marker = 'o', cmap='winter',label ='Actual Placements')
plt.scatter(df['CGPA'], df['Resume_Score'], c=predictions, marker = '*', label ='Predicted Placements')

plt.xlabel('CGPA')
plt.ylabel('Resume_Score')
plt.legend()
plt.show()
```
